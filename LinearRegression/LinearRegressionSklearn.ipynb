Hi! In this notebook, I'm implementing Linear Regression using **Sklearn**!

**1-** Importing necessary libraries:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import mplcyberpunk                        # an optional library for modernizing matplotlib
                                           # install : pip install mplcyberpunk
```

**2-** Loading the dataset and and separating it:

```python
df = pd.read_csv("LinearRegressionDatabase.csv")
X_train = df.Temperature.values.reshape(-1,1)
Y_train = df.Sales.values.reshape(-1,1)
```

**3-** Visualizing the dataset:

```python
plt.style.use('cyberpunk')
plt.scatter(X_train,Y_train)
mplcyberpunk.make_scatter_glow()
plt.xlabel('Temperature')
plt.ylabel('Sales')
plt.show()
```

**4-** Instantiating the model, provided by **sklearn**:

```python
regressor = LinearRegression()
```

**5-** Training the model:

```python
regressor.fit(X_train,Y_train)
```

**6-** Visualizing the final output:

```python
plt.scatter(X_train,Y_train)
plt.xlabel('Temperature')
plt.ylabel('Sales')
X_test = np.arange(X_train.min(), X_train.max(), 0.25).reshape(-1,1)
Y_test = regressor.predict(X_test)
plt.plot(X_test,Y_test,color='yellow')
mplcyberpunk.make_lines_glow()
mplcyberpunk.make_scatter_glow()
mplcyberpunk.add_underglow()
plt.show()
```
